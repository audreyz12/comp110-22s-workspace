### Part 4. Analysis

We begin by changing some settings in the notebook to automatically reload changes to imported files.

```python
%reload_ext autoreload
%autoreload 2
```

We continue by importing the helper functions from `data_utils`.

```python
from data_utils import read_csv_rows, columnar, head, select, count
```

Next, we create a path to the data and convert it to a 'table.' We can use `columnar` to convert this table to a column-oriented representation of the data. 

```python
SURVEY_DATA_CSV_FILE_PATH: str = "../../data/survey.csv"
data_rows: list[dict[str, str]] = read_csv_rows(SURVEY_DATA_CSV_FILE_PATH)
data_cols: dict[str, list[str]] = columnar(data_rows)
data_cols
```

We can use `head` to get an idea of what we're working with. 

```python
head(data_cols, 10)
```

That's a lot of factors to consider. Because we want to focus on the pace of teaching, we can look at the data column from the survey that directly corresponds with pace. We can achieve this with `select`.

```python
pace_cols: dict[str, list[str]] = select(data_cols, ["pace"])
pace_cols
```

We can use `count` to get a general idea of how students are feeling about the pace.

```python
pace_counts: dict[str, int] = count(pace_cols["pace"])
print(f"pace_counts: {pace_counts}")
```

To figure out the proportion of students that are finding the pace to be too fast, we can first figure out the total amount of students that responded to the survey with `add`.

```python
from data_utils import add
total: int = add(pace_counts)
total
```

We can look at the proportion of students for each response and add them into a new dictionary, for ease when it comes to understanding the data. There's definitely a larger amount of people that agree that the course is moving very quickly. The majority of people, about 32.7%, said 5. 

```python
from data_utils import proportions
proportion_cols: dict[str, float] = proportions(pace_counts, total)
proportion_cols
```

We can sort the proportions to make them easier to visualize. Let's create a bar chart of the information. 

```python
items_in_dict: list[tuple[str, float]] = list(proportion_cols.items())
items_in_dict = sorted(items_in_dict)
sorted_proportions: dict[str, float] = dict(items_in_dict)
sorted_proportions
```

```python
from matplotlib import pyplot

pyplot.title("Proportions of Pace Responses")
pyplot.xlabel("Responses")
pyplot.ylabel("Proportions")
labels: list[str] = list(sorted_proportions.keys())
values: list[float] = list(sorted_proportions.values())
pyplot.bar(labels, values)
```
